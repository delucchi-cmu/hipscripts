```python
import pandas as pd
```

```python

path="/data3/epyc/data3/hipscat/raw/pan_starrs/otmo/OTMO_20.csv"

type_frame = pd.read_csv("/data3/epyc/data3/hipscat/raw/pan_starrs/ps_otmo_types.csv")
type_names = type_frame["name"].values.tolist()
type_map = dict(zip(type_frame["name"], type_frame["type"]))

data_frame = pd.read_csv(path, index_col=False, header=None, nrows=10, 
                      names=type_names, dtype=type_map)

print(data_frame)
```
