```python
import healpy as hp
import numpy as np


catalog_path = ""
ra = 24.7035278
dec = -9.3653083
radius = 2 # arcsec

## What partitions/tiles/files are in this cone search?


highest_order = 10
nside = hp.order2nside(highest_order)
vec = hp.ang2vec(ra, dec, lonlat=True)
rad = np.radians(radius)
pixels_to_query = hp.query_disc(nside, vec, rad, nest=True)
pixels_to_query


order_map = np.full(hp.order2npix(10), hp.pixelfunc.UNSEEN)
order_map[pixels_to_query] = 4

hp.mollview(order_map, title="foo", nest=True)
```
