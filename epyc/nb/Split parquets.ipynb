```python
import glob
import glob
import os
import time

import pyarrow as pa
import pyarrow.parquet as pq

id_column = 'ps1_objid'
```

```python
start = time.perf_counter()

for file_index in range(100, 500):
    input_pattern = f'/data/epyc/projects/lsd2/pzwarehouse/ztf_dr14/part-{str(file_index).zfill(5)}*'
    
    input_paths = glob.glob(input_pattern)
    if(len(input_paths) != 1):
        print(f'BAD PATHS at {input_pattern}')
        continue
    
    parquet_file = pq.read_table(input_paths[0])
    parquet_file = parquet_file.sort_by(id_column)

    for chunk_index, smaller_table in enumerate(
            parquet_file.to_batches(max_chunksize=250_000)
        ):
        shard_path = f'/data3/epyc/data3/hipscat/ztf_shards/part-{str(file_index).zfill(5)}-shard-{chunk_index}.parquet'
        pq.write_table(pa.Table.from_batches([smaller_table]), shard_path)        
          
    end = time.perf_counter()
    print(f'finished file {file_index} in {int(end-start)} s')
    start = end
```
